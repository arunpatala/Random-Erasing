```python
def plot_imgs(Z):
    H = Z.shape[0]
    W = Z.shape[1]
    fig = plt.figure(figsize=(W, H))

    for i in range(H):
        for j in range(W):
            fig.add_subplot(H, W, i*W+1+j)
            plt.axis('off')   
            plt.imshow(Z[i][j], cmap='gray')
    plt.show()
```
